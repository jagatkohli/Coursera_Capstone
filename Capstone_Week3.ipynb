#### Get Postal Codes into Dataframe

```python
d = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df = d[0]
```

#### Dropping "Not assigned" values for Burough

```python
df_clean = df[df.Borough != "Not assigned"].reset_index(drop=True)
df_clean.head()
```

#### Grouping by Postal Code

```python
df2 = df_clean.groupby(['Postal Code', 'Borough'])['Neighborhood'].aggregate(lambda x: ', '.join(set(x))).reset_index()
df2.head()
```

#### Changing "Not assigned" Neigborhoods to their corresponding Boroughs

```python
for i, row in df2.iterrows():
    if row["Neighborhood"] == "Not assigned":
        row["Neighborhood"] = row["Borough"]
        
df2.head()
```

```python
df2.shape
```

#### Question 1 Solution

```python
solution = df2[df2['Postal Code'].isin(["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"])]
q1 = solution.iloc[pd.Index(solution['Postal Code']).get_indexer(["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"])]
q1.reset_index(drop=True)
```

#### Append postal codes with corresponding coordinates 

```python
df3 = df2.join(lat_long.set_index('Postal Code'), on='Postal Code')
df3.head()
```

#### Visualizing Toronto Postal Zones

```python
toronto_map = folium.Map(location=[43.6532, -79.3832], zoom_start=10.4)

feature = folium.map.FeatureGroup()

for lat, lng, in zip(df3.Latitude, df3.Longitude):
    feature.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5,
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add feature to map
toronto_map.add_child(feature)
```

#### Part 3 Solution: Clustering Postal Codes with Boroughs that contain "Toronto"

```python
df4 = df3[df3.Borough.str.contains('Toronto',case=False)]
df4.head()
```

```python
q3 = folium.map.FeatureGroup()

for lat, lng, in zip(df4.Latitude, df4.Longitude):
    q3.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5,
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add feature to map
toronto_map.add_child(q3)
```
